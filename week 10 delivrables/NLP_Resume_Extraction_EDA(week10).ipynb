```python
import matplotlib.pyplot as plt
import pandas as pd
import json
with open('/content/drive/MyDrive/NLP-DataGlacier/Resume.json') as f:
    data = [json.loads(line) for line in f]
df1=[]
df2=[]
for i in range(200):

  for j in range(len(data[i]["annotation"])):

    df1.append(data[i]["annotation"][j]["label"])
    x=data[i]["annotation"][j]["points"][0]["text"].lower()
    x=x.replace("\n", " ")
  
    df2.append(x)
df=pd.DataFrame(df1)
```

```python
df["text"]=df2
df
```

```python
L=[]
K=["oracle","microsoft","infosys","sap","cisco","accenture","amazon"]
for i in K:
  a=Companies[Companies["text"]==i].shape[0]
  L.append(int(a))
L
```

```python
x=0
for i in range(len(L)):
  x=x+L[i]
for i in range(len(L)):
  L[i]=(L[i]*x)/100
```

```python
labels =K
sizes = L

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title("Companies with the largest number of condidates worked at ")
plt.show()
```

```python
#number of engineers among the condidates
x=0
for i in range(268):
  B=Degree["text"]
  A=list(B)
  if((A[i].find("engineering")>=0) or (A[i].find("engineer")>=0)) :
      x=x+1
x
```

```python
#number of Computer scientists among the condidates
x=0
for i in range(268):
  B=Degree["text"]
  A=list(B)
  if( (A[i].find("information")>=0) or (A[i].find("computer")>=0) or (A[i].find("informatics")>=0)) :
      x=x+1
x
```

```python
#number of condidates with a master degree
x=0
for i in range(268):
  B=Degree["text"]
  A=list(B)
  if( (A[i].find("master")>=0) or (A[i].find("mba")>=0) ) :
      x=x+1
x
```

```python
#number of condidates with a bachelor degree
x=0
for i in range(268):
  B=Degree["text"]
  A=list(B)
  if( (A[i].find("bachelor")>=0) or (A[i].find("b.")>=0) or (A[i].find("bsc")>=0)) :
      x=x+1
x
```
